```python
from datetime import datetime, date, timedelta
```

**GETTING CURRENT DATE AND TIME**

```python
now = datetime.now()
print('Date:', now.date())
print('TIme:', now.time())
```

```python
print(date.today())
```

**CUSTOM FORMATTING**

```python
print(now.strftime("%d-%m-%Y %H:%M:%S"))
```

```python
print(now.strftime("%Y-%m-%d %H:%M:%S"))
```

**SPECIFIC DATES/TIMES**

```python
dt = datetime(2024, 12, 31, 23, 59, 59)
print(dt)
```

**TIMEDELTA**

```python
start = datetime.now()

#some code
for i in range(100):
    print(i)

end = datetime.now()

duration = end - start

print(duration)
```

```python
tomorrow = datetime.now() + timedelta(hours = 1)
print(tomorrow)
```

**WORKING ONLY WITH DATES**

```python
now = datetime.now()
print("Today:", now)
print("Year:", now.year)
print("Month:", now.month)
print("Day:", now.day)
print("Hour:", now.hour)
```

**PARSE STRING -> DATETIME**

```python
# Convert a text timestamp (common in datasets) into a datetime object.
dt = datetime.strptime("2025-11-02 22:45", "%Y-%m-%d %H:%M")
print(dt)
```
