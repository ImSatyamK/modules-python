```python
import math
```

**BASIC**

```python
math.sqrt(125) #square root
```

```python
math.pow(50,2) #power
```

```python
math.exp(2) #e to the power of 2
```

```python
math.fabs(-56) #absolute value in float
```

**ROUNDING**

```python
math.floor(5.99) #rounding down
```

```python
math.ceil(5.01) #rounding up
```

```python
math.trunc(5.99) #removes decimal part
```

**CONSTANTS**

```python
math.pi  #value of pi
```

```python
math.e  #value of e
```

**TRIGNOMETRY**

```python
math.sin(math.pi/2)  #sine of 90 degrees
```

```python
math.cos(math.pi)  #cosine of 180 degrees
```

```python
math.tan(math.pi/4)  #tangent of 45 degrees
```

**CONVERSION**

```python
math.degrees(math.pi/2)  #radians to degrees
```

```python
math.radians(90)  #degrees to radians
```

**LOGARITHMS**

```python
math.log(100)  #natural logarithm with base e
```

```python
math.log10(1000)  #logarithm with base 10
```

```python
math.log2(1024)  #logarithm with base 2
```

**FACTORIAL AND GCD**

```python
math.factorial(5)  #factorial of 5
```

```python
math.gcd(48, 180)  #greatest common divisor
```

```python
math.lcm(12, 15)  #least common multiple
```

**PERMUTATIONS AND COMBINATION**

```python
math.perm(5, 2)  #permutations of 5 items taken 2 at a time
```

```python
math.comb(5, 2)  #combinations of 5 items taken 2 at a time
```

**PRECISION UTILITIES**

```python
values = [0.1, 0.1, 0.1]
print(sum(values))  #simple sum of floating-point numbers
print(math.fsum(values))  #precise sum of floating-point numbers
```
